```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
from geopy.geocoders import Nominatim

# Step 1: Read the CSV file into a DataFrame
df = pd.read_csv('Database.csv')

# Step 2: Create a new DataFrame to store the individual countries and their counts
country_counts = pd.DataFrame(columns=['Country', 'Count'])

# Step 3: Loop through the "Study_location" column and count each country
for locations in df['Study_location']:
    # Check if the locations are not NaN (i.e., valid string)
    if isinstance(locations, str):
        # Split the locations by comma and strip any leading/trailing whitespaces
        countries = [country.strip() for country in locations.split(',')]
        
        # Update the count for each country in the country_counts DataFrame
        for country in countries:
            if country:
                # Check if the country already exists in the DataFrame
                if country in country_counts['Country'].values:
                    # Increment the count for existing country
                    country_counts.loc[country_counts['Country'] == country, 'Count'] += 1
                else:
                    # Add the country with count 1 if it doesn't exist in the DataFrame
                    new_row = pd.DataFrame({'Country': [country], 'Count': [1]})
                    country_counts = pd.concat([country_counts, new_row], ignore_index=True)

# Step 4: Calculate the total count of countries
total_countries = country_counts['Count'].sum()

# Step 5: Calculate the percentage for each country
country_counts['Percentage'] = (country_counts['Count'] / total_countries) * 100

# Step 6: Sort the DataFrame in descending order based on the count
country_counts = country_counts.sort_values(by='Count', ascending=False)

# Step 7: Print the results (individual country counts and percentages)
print("Individual Country Counts:")
print(country_counts[['Country', 'Count']])

print("\nIndividual Country Percentages:")
print(country_counts[['Country', 'Percentage']])

# Step 8: Obtain the latitude and longitude for each country
geolocator = Nominatim(user_agent="myGeocoder")
country_counts['Location'] = country_counts['Country'].apply(lambda country: geolocator.geocode(country))

# Step 9: Filter out the countries for which location could not be found
country_counts = country_counts.dropna(subset=['Location'])

# Step 10: Create a GeoDataFrame to store the locations
gdf = gpd.GeoDataFrame(country_counts, geometry=gpd.points_from_xy(country_counts['Location'].apply(lambda loc: loc.longitude), country_counts['Location'].apply(lambda loc: loc.latitude)))

# Step 11: Load the world map shapefile
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Step 12: Plot the world map
fig, ax = plt.subplots(figsize=(15, 10))
world.plot(ax=ax, color='lightgrey', edgecolor='black')

# Step 13: Plot the locations on the map
gdf.plot(ax=ax, markersize='Count', color='red', marker='o', label='Locations')

# Step 14: Set title and legend
ax.set_title('Study Locations on World Map')
ax.legend()

# Step 15: Show the map
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
from geopy.geocoders import Nominatim
from matplotlib.path import Path
import matplotlib.patches as patches

# Step 1: Read the CSV file into a DataFrame
df = pd.read_csv('Database.csv')

# Step 2: Create a new DataFrame to store the individual countries and their counts
country_counts = pd.DataFrame(columns=['Country', 'Count'])

# Step 3: Loop through the "Study_location" column and count each country
for locations in df['Study_location']:
    # Check if the locations are not NaN (i.e., valid string)
    if isinstance(locations, str):
        # Split the locations by comma and strip any leading/trailing whitespaces
        countries = [country.strip() for country in locations.split(',')]
        
        # Update the count for each country in the country_counts DataFrame
        for country in countries:
            if country:
                # Check if the country already exists in the DataFrame
                if country in country_counts['Country'].values:
                    # Increment the count for existing country
                    country_counts.loc[country_counts['Country'] == country, 'Count'] += 1
                else:
                    # Add the country with count 1 if it doesn't exist in the DataFrame
                    new_row = pd.DataFrame({'Country': [country], 'Count': [1]})
                    country_counts = pd.concat([country_counts, new_row], ignore_index=True)

# Step 4: Calculate the total count of countries
total_countries = country_counts['Count'].sum()

# Step 5: Calculate the percentage for each country
country_counts['Percentage'] = (country_counts['Count'] / total_countries) * 100

# Step 6: Sort the DataFrame in descending order based on the count
country_counts = country_counts.sort_values(by='Count', ascending=False)

# Step 7: Print the results (individual country counts and percentages)
print("Individual Country Counts:")
print(country_counts[['Country', 'Count']])

print("\nIndividual Country Percentages:")
print(country_counts[['Country', 'Percentage']])

# Step 8: Obtain the latitude and longitude for each country
geolocator = Nominatim(user_agent="myGeocoder")
country_counts['Location'] = country_counts['Country'].apply(lambda country: geolocator.geocode(country))

# Step 9: Filter out the countries for which location could not be found
country_counts = country_counts.dropna(subset=['Location'])

# Step 10: Convert the "Count" column to a numeric data type
country_counts['Count'] = pd.to_numeric(country_counts['Count'])

# Step 11: Create a GeoDataFrame to store the locations
gdf = gpd.GeoDataFrame(country_counts, geometry=gpd.points_from_xy(country_counts['Location'].apply(lambda loc: loc.longitude), country_counts['Location'].apply(lambda loc: loc.latitude)))

# Step 12: Load the world map shapefile
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Step 13: Plot the world map
fig, ax = plt.subplots(figsize=(15, 10))
world.plot(ax=ax, color='lightgrey', edgecolor='black')

# Step 14: Plot the locations on the map with the custom location sign marker
for idx, row in gdf.iterrows():
    x, y = row.geometry.x, row.geometry.y
    path_data = [
        (Path.MOVETO, (x, y + 0.1)),
        (Path.LINETO, (x + 0.05, y)),
        (Path.LINETO, (x, y - 0.1)),
        (Path.LINETO, (x - 0.05, y)),
        (Path.CLOSEPOLY, (x, y + 0.1))
    ]
    codes, verts = zip(*path_data)
    path = Path(verts, codes)
    patch = patches.PathPatch(path, facecolor='red', edgecolor='black', lw=1.5)
    ax.add_patch(patch)

# Step 15: Set title and legend
ax.set_title('Study Locations on World Map')
ax.legend()

# Step 16: Show the map
plt.show()
```

```python
import pandas as pd
import folium
from geopy.geocoders import Nominatim

# Step 1: Read the CSV file into a DataFrame
df = pd.read_csv('Database.csv')

# Step 2: Create a new DataFrame to store the individual countries and their counts
country_counts = pd.DataFrame(columns=['Country', 'Count'])

# Step 3: Loop through the "Study_location" column and count each country
for locations in df['Study_location']:
    # Check if the locations are not NaN (i.e., valid string)
    if isinstance(locations, str):
        # Split the locations by comma and strip any leading/trailing whitespaces
        countries = [country.strip() for country in locations.split(',')]
        
        # Update the count for each country in the country_counts DataFrame
        for country in countries:
            if country:
                # Check if the country already exists in the DataFrame
                if country in country_counts['Country'].values:
                    # Increment the count for existing country
                    country_counts.loc[country_counts['Country'] == country, 'Count'] += 1
                else:
                    # Add the country with count 1 if it doesn't exist in the DataFrame
                    new_row = pd.DataFrame({'Country': [country], 'Count': [1]})
                    country_counts = pd.concat([country_counts, new_row], ignore_index=True)

# Step 4: Calculate the total count of countries
total_countries = country_counts['Count'].sum()

# Step 5: Calculate the percentage for each country
country_counts['Percentage'] = (country_counts['Count'] / total_countries) * 100

# Step 6: Sort the DataFrame in descending order based on the count
country_counts = country_counts.sort_values(by='Count', ascending=False)

# Step 7: Print the results (individual country counts and percentages)
print("Individual Country Counts:")
print(country_counts[['Country', 'Count']])

print("\nIndividual Country Percentages:")
print(country_counts[['Country', 'Percentage']])

# Step 8: Manually set the latitude and longitude for Brazil
brazil_latitude = -14.2350
brazil_longitude = -51.9253

# Step 9: Create a folium Map
m = folium.Map(location=[0, 0], zoom_start=2)

# Step 10: Obtain the latitude and longitude for each country
geolocator = Nominatim(user_agent="myGeocoder")
for idx, row in country_counts.iterrows():
    location = None
    if row['Country'] == 'Brazil':
        # If the country is Brazil, use the manually set latitude and longitude
        location = [brazil_latitude, brazil_longitude]
    else:
        # For other countries, use the geocoded latitude and longitude
        location_info = geolocator.geocode(row['Country'])
        if location_info:
            location = [location_info.latitude, location_info.longitude]

    if location:
        folium.Marker(
            location=location,
            popup=f"{row['Country']} ({row['Count']} studies)",
            icon=folium.Icon(icon='info-sign', color='red')
        ).add_to(m)

# Step 11: Display the map
m.save('study_locations_map.html')

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
file_path = "Database.csv"  # Replace with the actual path to your CSV file
df = pd.read_csv(file_path)

# Assuming the column name is "Year", you can access it like this:
year_column = df["Year"]

# Drop rows with invalid "Year" values (NaN or inf)
year_column = year_column.dropna().astype(int)

# Calculate the frequency of each unique year value in the "Year" column
year_counts = year_column.value_counts()

# Add year 2023 with count 0 if it doesn't exist in the data
year_counts = year_counts.reindex(range(year_column.min(), year_column.max() + 1), fill_value=0)

# Create the line chart
plt.figure(figsize=(10, 6))  # Set the figure size (width, height)

# Plot the line chart
plt.plot(year_counts.index, year_counts.values, marker='o', linestyle='-')

# Set labels and title
plt.xlabel("Year")
plt.ylabel("Count")
plt.title("Count of Years in Database")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the grid
plt.grid()

# Display the line chart
plt.show()
```
